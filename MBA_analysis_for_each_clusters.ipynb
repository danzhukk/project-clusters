MBA analysis for cluster 0

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./datasets/Data_for_MBA_all_clusters_true.csv')
df.set_index('id', inplace = True)
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df_0 = df.loc[df['Cluster'] == 0]
```

```python
df_0.shape
```

```python
MarkBas_0 = df_0.drop(columns='Cluster', axis=1)
```

```python
MarkBas_0.head()
```

```python
def encode_values(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_encoded_0 = MarkBas_0.map(encode_values)

basket_encoded_0.head()
```

```python
basket_filtered_0 = basket_encoded_0[(basket_encoded_0 > 0).sum(axis=1) >= 2]
basket_filtered_0.head()
```

```python
basket_filtered_0.shape
```

```python
basket_bool = basket_filtered_0.astype(bool)

frequent_itemsets_0 = apriori(
    basket_bool, 
    min_support=0.03, 
    use_colnames=True
).sort_values("support", ascending=False)

frequent_itemsets_0.head(10)
```

```python
assoc_rules_0 = association_rules(frequent_itemsets_0, metric="lift", min_threshold=1.1)
assoc_rules_0 = assoc_rules_0[assoc_rules_0['consequents'].apply(len) == 1]
assoc_rules_0 = assoc_rules_0[assoc_rules_0['lift'] > 1]
assoc_rules_0 = assoc_rules_0.sort_values("lift", ascending=False).reset_index(drop=True)
assoc_rules_0.head()
```

```python
df_with_price = pd.read_csv('./datasets/df_with_price.csv')
df_with_price.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df_with_price.head()
```

```python
margin_map = (
    df_with_price.groupby('Группа1_Группа2')['Маржа']
    .mean()
    .reset_index(name='avg_margin')
    .set_index('Группа1_Группа2')['avg_margin']
)

assoc_rules_0['consequent_item'] = assoc_rules_0['consequents'].apply(lambda x: next(iter(x)))

assoc_rules_0['avg_margin_consequent'] = assoc_rules_0['consequent_item'].map(margin_map)

assoc_rules_0 = assoc_rules_0[assoc_rules_0['avg_margin_consequent'] > 0]


# 4. Рассчитываем новый profit_score
assoc_rules_0['profit_score'] = (
    assoc_rules_0['confidence'] 
    * assoc_rules_0['lift'] 
    * assoc_rules_0['avg_margin_consequent']
)

# 5. Сортируем по profit_score
assoc_rules_0 = assoc_rules_0.sort_values('profit_score', ascending=False)

# 6. Выводим топ-рекомендаций
assoc_rules_0[['antecedents', 'consequents', 'confidence', 'lift', 'avg_margin_consequent', 'profit_score']].head(10)
```

```python
valid_phones = df_0.index.unique()
df_cluster0 = df_with_price[df_with_price['Телефон_new'].isin(valid_phones)]
customer_items = df_cluster0.groupby('Телефон_new')['Группа1_Группа2'].apply(set)

recs = []
for _, rule in assoc_rules_0.iterrows():
    A = set(rule['antecedents'])
    C = rule['consequent_item']
    for phone, items in customer_items.items():
        if A.issubset(items) and (C not in items):
            recs.append({
                'Телефон_new': phone,
                'antecedents': ", ".join(rule['antecedents']),
                'recommend_item': C,
                'confidence': rule['confidence'],
                'lift': rule['lift'],
                'avg_margin': rule['avg_margin_consequent'],
                'profit_score': rule['profit_score']
            })

recommend_df_0 = pd.DataFrame(recs)
recommend_df_0 = recommend_df_0.sort_values('profit_score', ascending=False)
recommend_df_0 = recommend_df_0.drop_duplicates(['Телефон_new', 'recommend_item'])

recommend_df_0.head()
```
